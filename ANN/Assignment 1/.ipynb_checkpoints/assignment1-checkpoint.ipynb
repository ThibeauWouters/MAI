# Approximating a sine wave

## Discussing speed and general performance

```python
speed_df = pd.read_csv("Data/sine_speed_results.csv")#.groupby("Algorithm").mean()
speed_df
```

```python
algorithms = speed_df["Algorithm"].unique()
xlabels = algorithms
print(xlabels)
colors = ["blue", "gray", "orange", "red", "green", "purple"]
xlabels = ["GD", "GDX", "LM", "CGF", "BFG", "BR"]
```

```python
plt.figure(figsize=(7,3))
# Define the plot parameters for the loop
key = "Training Time"

# Save data of each algorithm and for both pure and noisy data
# this_df = df_n[(df_n["Iterations"] == np.max(df_n["Iterations"])) & (df_n["Hidden"] == np.max(df_n["Hidden"]))]
# this_df = df[(df["Iterations"] == np.max(df["Iterations"])) & (df["Hidden"] == np.max(df["Hidden"]))]
# mu = []
# sigma = []
for i, alg in enumerate(algorithms):
    data = speed_df[speed_df["Algorithm"]     == alg][key]
    # Make plot
    plt.errorbar(i+1, np.mean(data), yerr=np.std(data), fmt = 'o', capsize = 7, color = colors[i])

# Set the x-axis tick labels and the axis label
xticks = [i+1 for i in range(len(algorithms))]
plt.xticks(xticks, xlabels, fontsize=fs)

# Set the y-axis label
plt.ylabel("Training time (s)", fontsize=fs)
# plt.yscale('log')
plt.grid()
plt.yticks(fontsize=fs)
plt.savefig("Plots/ex1_sine_training_time_final.pdf", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(7,3))
# Define the plot parameters for the loop
key = "Train MSE"

# Save data of each algorithm and for both pure and noisy data
# this_df = df_n[(df_n["Iterations"] == np.max(df_n["Iterations"])) & (df_n["Hidden"] == np.max(df_n["Hidden"]))]
# this_df = df[(df["Iterations"] == np.max(df["Iterations"])) & (df["Hidden"] == np.max(df["Hidden"]))]
# mu = []
# sigma = []
for i, alg in enumerate(algorithms):
    data = speed_df[speed_df["Algorithm"]     == alg][key]
    # Make plot
    plt.errorbar(i+1, np.mean(data), yerr=np.std(data), fmt = 'o', capsize = 7, color = colors[i])

# Set the x-axis tick labels and the axis label
xticks = [i+1 for i in range(len(algorithms))]
plt.xticks(xticks, xlabels, fontsize=fs)

# Set the y-axis label
plt.ylabel(key, fontsize=fs)
plt.yscale('log')
plt.grid()
plt.yticks(fontsize=fs)
plt.savefig("Plots/ex1_sine_performance_final.pdf", bbox_inches='tight')
plt.show()
```

## Generalization and noise

```python
df = pd.read_csv(os.path.join(data_dir, "sine_results_no_noise_FINAL.csv"))#.groupby(["Hidden", "Algorithm"]).mean()
# df_n    = pd.read_csv(os.path.join(data_dir, "sine_results_noise_FINAL.csv"))
df_n    = pd.read_csv(os.path.join(data_dir, "sine_results_noise_FINAL_0.2.csv"))
hidden_list = np.unique(df["Hidden"].values)
# Show an example
# df.head(10)
```

```python
# # Switch data
# df = pd.read_csv(os.path.join(data_dir, "generalization_no_noise.csv"))
# df_n = pd.read_csv(os.path.join(data_dir, "generalization_noise.csv"))
# # Show an example
# df.head(3)
```

This is to get df for one algorithm:

```python
# # df[df["Algorithm"] == "traingd"]
# this_df = df[df["Algorithm"] == "traingd"].groupby(["Hidden"]).mean()
# this_df
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(11,2), sharey=True)
ax1, ax2 = axs
keys = ["Train MSE", "Test MSE"]

# Make plots:
for a, ax in enumerate(axs):
    key = keys[a]
    for i, alg in enumerate(algorithms):
        this_df = df[df["Algorithm"] == alg].groupby("Hidden").mean()
        train_vals = this_df[key].values
        ax.plot(hidden_list, train_vals, '-o', ms=3, color = colors[i], label = xlabels[i])
        #ax.set_yticks(fontsize=fs)
        ax.set_xlabel("Hidden layer size", fontsize = fs)
        ax.set_yscale('log')
        
# Set labels, titles, grids:
ax1.set_ylabel("MSE", fontsize=fs)
ax1.set_title("Train MSE  (no noise)", fontsize=fs)

ax2.set_title("Test MSE  (no noise)", fontsize=fs)
for ax in axs:
    ax.grid()
        
plt.legend(fontsize=fs-3, bbox_to_anchor=(1.00, 1.05))
plt.savefig("Plots/ex1_sine_generalization.pdf", bbox_inches='tight')
plt.show()
```

Now, with noise

```python
ncols = 3

fig, axs = plt.subplots(nrows=1, ncols=ncols, figsize=(11,2), sharey=True)
if ncols == 3:
    ax1, ax2, ax3 = axs
else:
    ax1, ax2 = axs
keys = ["Train MSE", "Test MSE noise", "Test MSE"]
titles = ["Train MSE", r"Test MSE $y$", r"Test MSE $\tilde{y}$"]

# Make plots:
for a, ax in enumerate(axs):
    key = keys[a]
    for i, alg in enumerate(algorithms):
        this_df = df_n[df_n["Algorithm"] == alg].groupby("Hidden").mean()
        train_vals = this_df[key].values
        ax.plot(hidden_list, train_vals, '-o', ms=3, color = colors[i], label = xlabels[i])
        #ax.set_yticks(fontsize=fs)
        ax.set_xlabel("Hidden layer size", fontsize = fs)
        ax.set_yscale('log')
        ax.set_title(titles[a])
        
# Set labels, titles, grids:
ax1.set_ylabel("MSE", fontsize=fs)
# ax1.set_title("Train MSE  (noise)", fontsize=fs)

# ax2.set_title("Test MSE  (noise)", fontsize=fs)
for ax in axs:
    ax.grid()
        
plt.legend(fontsize=fs-3, bbox_to_anchor=(1.00, 1.05))
plt.savefig("Plots/ex1_sine_generalization_noise.pdf", bbox_inches='tight')
plt.show()
```

```python
xvals = np.loadtxt("Data/xvals.csv", delimiter=",")
sine = np.loadtxt("Data/sine_simulated.csv", delimiter=",")
sine_noise = np.loadtxt("Data/sine_simulated_noise.csv", delimiter=",")
true = np.sin(xvals**2)

plt.figure(figsize=(11,3))

plt.plot(xvals, true, "-", color="gray", alpha = 0.75, label="True")
plt.plot(xvals, sine, "-", color="blue", alpha = 0.75, label="No noise")
plt.plot(xvals, sine_noise, "-", color="red", alpha = 0.75, label="Noise")

plt.grid()
plt.xlabel(r"$x$", fontsize=fs)
plt.ylabel(r"$y$", fontsize=fs)
plt.legend()
plt.savefig("Plots/ex1_sine_visual_80.pdf", bbox_inches='tight')
plt.show()
```

Read in our samples of the final prediction after tuning

```python
train_input = np.loadtxt("Data/personal_train_input.txt", delimiter=",")
train_target = np.loadtxt("Data/personal_train_target.txt", delimiter=",")
test_input = np.loadtxt("Data/personal_test_input.txt", delimiter=",")
test_target = np.loadtxt("Data/personal_test_target.txt", delimiter=",")
preds = np.loadtxt("Data/personal_test_preds.txt", delimiter=",")
```

```python
X1, X2 = train_input
Tnew = train_target
Tnew.shape
```

Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, Tnew, c=Tnew)

# Set labels for the axes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('T')

# Show the plot
plt.show()
```

Now, with an interpolant:

```python
from scipy.interpolate import Rbf
```

```python
# Define a grid of X1 and X2 values
xi = np.linspace(X1.min(), X1.max(), 100)
yi = np.linspace(X2.min(), X2.max(), 100)
xi, yi = np.meshgrid(xi, yi)

# Use Rbf to create an interpolant function
rbf = Rbf(X1, X2, Tnew, function='linear')
zi = rbf(xi, yi)

# Create a 3D plot of the smoothed surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xi, yi, zi, cmap='viridis', alpha=1)
# Set labels for the axes
ax.set_xlabel(r'$X_1$', fontsize=fs)
ax.set_ylabel(r'$X_2$', fontsize=fs)
ax.set_zlabel(r'$T$', fontsize=fs)

# Show the plot
plt.savefig("Plots/personal_train_surface.pdf", bbox_inches = 'tight')
plt.show()
```

## Performance of the algorithms

```python
import seaborn as sn
from  matplotlib.colors import LinearSegmentedColormap
import matplotlib.colors as mcolors

my_cmap=LinearSegmentedColormap.from_list('rg',["r", "w", "g"], N=256) 
my_cmap_r=LinearSegmentedColormap.from_list('rg',["g", "w", "r"], N=256) 
```

```python
df = pd.read_csv("Data/personal_regression_avg.csv")
mean_df = df.groupby(["Num layers", "Num hidden"]).mean()
mean_df
```

```python
mse_values = mean_df["Val MSE"].values
mse_values
```

```python
# Reshape into a matrix for plotting
layers_values = np.unique(df["Num layers"].values)
hidden_values = np.unique(df["Num hidden"].values)
mse_values = mse_values.reshape((len(layers_values), len(hidden_values)))
```

```python
plt.figure(figsize = (8,2))
df_for_plot = pd.DataFrame(mse_values, index = layers_values, columns = hidden_values)
# plt.figure(figsize = (5,3))
sn.heatmap(df_for_plot, annot=True, cmap=my_cmap_r, linewidths=1, linecolor='black', norm=mcolors.LogNorm(), cbar_kws={'label': 'Validation MSE'})
plt.xlabel("Hidden size", fontsize=fs)
plt.ylabel("Number of layers", fontsize=fs)
# plt.title("Validation MSE")
plt.savefig("Plots/ex2_gridsearch.pdf", bbox_inches = "tight")
plt.show()
```

## Test set and errors

```python
X1, X2 = test_input
Tnew = test_target
That = preds
That.shape
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a grid of X1 and X2 values
xi = np.linspace(X1.min(), X1.max(), 100)
yi = np.linspace(X2.min(), X2.max(), 100)
xi, yi = np.meshgrid(xi, yi)

# Use Rbf to create an interpolant function
rbf = Rbf(X1, X2, Tnew, function='linear')
zi = rbf(xi, yi)

# Plot surface
ax.plot_surface(xi, yi, zi, cmap='viridis', alpha=0.25)

# Scatter the predictions
ax.scatter(X1, X2, preds, c=preds)

# Set labels for the axes
ax.set_xlabel(r'$X_1$', fontsize=fs)
ax.set_ylabel(r'$X_2$', fontsize=fs)
ax.set_zlabel(r'$T$', fontsize=fs)

# Show the plot
plt.savefig("Plots/personal_test_preds.pdf", bbox_inches = 'tight')
plt.show()
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X1, X2, abs(Tnew - preds), c=abs(Tnew - preds), cmap='coolwarm')

# Set labels for the axes
ax.set_xlabel(r'$X_1$', fontsize=fs)
ax.set_ylabel(r'$X_2$', fontsize=fs)
ax.set_zlabel('Absolute error', fontsize=fs, labelpad = 15)

# Show the plot
plt.savefig("Plots/personal_errors.pdf", bbox_inches='tight')
plt.show()
```
