```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1

```python
x1 = np.random.normal(size=(50, 2)) + 1
x2 = np.random.normal(size=(51, 2))  - 1

mean1 = np.mean(x1, axis = 0)
mean2 = np.mean(x2, axis = 0)

y1 = np.ones(50)
y2 = np.ones(51)
```

```python
center_x, center_y = (mean1+mean2)/2
center_x, center_y
```

Get the coordinates of the two means

```python
xc1, yc1 = mean1
xc2, yc2 = mean2
```

```python
m = (yc2 - yc1)/(xc2 - xc1)
m = -1/m
```

```python
def decision_boundary(x):
    
    return center_y + m*(x - center_x)
```

```python
plt.scatter(x1[:, 0], x1[:, 1], marker = 'o', color="red", label = "Class 1")
plt.scatter(x2[:, 0], x2[:, 1], marker = 'o', color="blue", label = "Class 2")
x_vals = np.arange(-4, 4, 0.0001)
y_vals = decision_boundary(x_vals)
plt.plot(x_vals, y_vals, '--', color = "black", label = "Decision boundary")
# plt.axhline(0, color="black")
# plt.axvline(0, color="black")
plt.xlabel("x")
plt.xlabel("y")
plt.legend()
plt.grid()
plt.show()
```
